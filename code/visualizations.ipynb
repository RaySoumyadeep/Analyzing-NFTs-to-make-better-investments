```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv(r"C:\Users\Neeraj Kankani\OneDrive - Oklahoma A and M System\Desktop\PDS2\msis-5223-deliverable-2-kidcodi\data\merged_data.csv")
data.columns
```

```python
dataset = data[data['total'] < 6000]
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.violinplot(x='tier', y='dollar-price', data=data)
ax.set_title("Dollar Price by Tiers")
ax.set_ylabel("Dollar-price")
ax.set_xlabel("Tiers of CoolCats")

plt.show()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.scatterplot(x='total', y='dollar-price', data=dataset)
ax.set_title("Dollar Price by Total score")
ax.set_ylabel("Dollar-price")
ax.set_xlabel("Total score based on rarity of CoolCats")

plt.show()
```

```python
dataset = data[data['dollar-price'] < 250000]
dataset = dataset[dataset['total'] < 2000]
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.histplot(x='rank', y='dollar-price', data=dataset, color='red', alpha = 1, bins=20)
ax.set_title("Dollar Price by Ranks")
ax.set_ylabel("Dollar-price")
ax.set_xlabel("Ranks of CoolCats")

plt.show()
```

```python
dataset = data[['dollar-price', 'rank', 'face-score','shirt-score', 'tier-score', 'hats-score','total', 'price']]
```

```python
sns.pairplot(dataset)
```

```python
# Plotting the KDE Plot

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2,figsize=(20,20))

sns.lineplot(x = data['face-score'], y = data['dollar-price'], color='blue', ax =ax1)

ax1.set_title("Face Score Distribution",font="Serif")



sns.lineplot(x = data['shirt-score'], y =  data['dollar-price'], color='green', ax =ax2)

ax2.set_title("Shirt Score Distribution",font="Serif")




sns.lineplot(x = data['tier-score'],  y = data['dollar-price'], color='red', ax =ax3)

ax3.set_title("Tier Score Distribution",font="Serif")




sns.lineplot(x = data['hats-score'], y = data['dollar-price'], color='purple', ax =ax4)

ax4.set_title("Hats Score Distribution",font="Serif")
```

```python

fig, ((ax5), (ax6)) = plt.subplots(2,1,figsize=(20,20))

sns.scatterplot(x = dataset['total'], y = dataset['dollar-price'], color='red', ax =ax5)

ax5.set_title("Total Score Distribution",font="Serif")




sns.scatterplot(x = dataset['rank'], y = dataset['dollar-price'], color='brown', ax =ax6)

ax6.set_title("Rank Distribution",font="Serif")
```

```python
# show the mean for each unique value of tier.
data.groupby('tier', as_index=False)['dollar-price'].mean()

sns.set(rc={'figure.figsize':(11.7,8.27)})

ax = sns.barplot(x = 'tier', y = 'dollar-price', data=data,ci=False)

ax.set_title("Average Dollar Price by Tiers")
ax.set_ylabel("Dollar-price")
ax.set_xlabel("Tiers of Coolcats")

plt.show()
```
